```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
car_sales=pd.read_csv("data/car-sales.csv")
car_sales_dataFrame=pd.DataFrame(car_sales)
print (car_sales_dataFrame)
```

```python
# Try to plot the 'Price' column using the plot() function
car_sales_dataFrame['Price'].plot()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales_dataFrame['Price'] = car_sales_dataFrame['Price'].str.replace('$', '')
car_sales_dataFrame['Price'] = car_sales_dataFrame['Price'].str.replace(',', '').astype(float)
```

```python
from datetime import date, timedelta
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column

car_sales_dataFrame['Total Sales'] = car_sales_dataFrame['Price'].cumsum()
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
start_date = date.today()
end_date = start_date + timedelta(days=len(car_sales_dataFrame))
date_range = pd.date_range(start=start_date, end=end_date, periods=len(car_sales_dataFrame))
car_sales_dataFrame['Sale Date'] = date_range

# View the car_sales DataFrame
print(car_sales_dataFrame)
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
plt.plot(car_sales_dataFrame['Sale Date'] , car_sales_dataFrame['Total Sales'])
plt.xlabel('Sale Date')
plt.ylabel('Total Sales ($)')
plt.title('Total Sales Over Time')
```

```python
# Convert the 'Price' column to the integers
car_sales_dataFrame['Total Sales'] =  car_sales_dataFrame['Total Sales'].astype(int)

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales_dataFrame.plot ( 'Odometer (KM)' , 'Price' , kind='scatter' )
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in **patients over 50 years old**.

For this next challenge, we're going to be replicating the following plot:

<img src="../images/matplotlib-heart-disease-chol-age-plot.png"/>

```python
# Replicate the above plot in whichever way you see fit
heart_disease['age'].plot.hist(subplots=True, figsize = (7 ,5)  ,edgecolor='black')
# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old

patients_over_50 = heart_disease[heart_disease['age'] > 50]


# Calculate the mean age and cholesterol for all patients over 50
mean_age = patients_over_50['age'].mean()
mean_chol = patients_over_50['chol'].mean()

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the data

#  no heart disease target=0
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 0]['age'],
    y=patients_over_50[patients_over_50['target'] == 0]['chol'],
    c='blue',
    label='No Heart Disease',
    alpha=0.7
)


#  with heart disease  target=1
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 1]['age'],
    y=patients_over_50[patients_over_50['target'] == 1]['chol'],
    c='red',
    label='Heart Disease',
    alpha=0.7
)


# Customize the plot

plt.title('Age vs. Cholesterol for Patients Over 50')
plt.xlabel('Age')
plt.ylabel('Cholesterol (chol)')
plt.legend()

# Add a meanline


# Add mean lines
plt.axvline(mean_age, color='green', linestyle='--', label=f'Mean Age: {mean_age:.2f}')
plt.axhline(mean_chol, color='y', linestyle='--', label=f'Mean Chol: {mean_chol:.2f}')

plt.legend()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like this:
<img src="../images/matplotlib-heart-disease-chol-age-plot-cmap-change.png"/>

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot
mean_age = patients_over_50['age'].mean()
mean_chol = patients_over_50['chol'].mean()

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the data for patients with no heart disease (target=0)
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 0]['age'],
    y=patients_over_50[patients_over_50['target'] == 0]['chol'],
    c='blue',
    label='No Heart Disease',
    alpha=0.7,
    cmap='winter'  # Change the cmap parameter to 'winter'
)

# Plot the data for patients with heart disease (target=1)
plt.scatter(
    x=patients_over_50[patients_over_50['target'] == 1]['age'],
    y=patients_over_50[patients_over_50['target'] == 1]['chol'],
    label='Heart Disease',
    alpha=0.7,
    cmap='winter'  # Change the cmap parameter to 'winter'
)

# Customize the plot
plt.title('Age vs. Cholesterol for Patients Over 50')
plt.xlabel('Age')
plt.ylabel('Cholesterol (chol)')
plt.legend()

# Add mean lines
plt.axvline(mean_age, color='red', linestyle='--', label=f'Mean Age: {mean_age:.2f}')
plt.axhline(mean_chol, color='red', linestyle='--', label=f'Mean Chol: {mean_chol:.2f}')  # Change the color to 'red'

# Save the plot to a file before showing
plt.savefig('heart_disease_plot.png')

# Show the plot
plt.legend()
plt.show()
```

Beautiful! Now our figure has an upgraded color scheme let's save it to file.
